#데이터 전처리
NAN은 lable0과 같은 종류이고, missing은 독립적인 종류임을 확인.

missing은 2번, NAN은 0번 할당

1. Label 재할당
2. Oversampling을 통한 데이터 불균형 해소
3. pixel값 Scaling
4. 학습을 위한 형태(keras의 경우 (sample,width,height,channel)로 입력)인 (28,28)로 reshape
5. Label을 LabelBinarizer로 변환(원-핫 인코딩 방식)

```python
train['label']=train['label'].apply(lambda x:'2' if x=='missing' else x)
train['label']=train['label'].apply(lambda x:'0' if x=='NAN' else x)
```

```python
y=train['label'].astype(int).values
X=train.drop(['id','label'],axis=1).values
```

#Model 정의하기
CNN모델을 활용하였으며, 28x28크기가 작다고 판단하여 먼저 channel을 늘려 데이터에서 추출할 수 있는 특징을 최대한 추출하려했음.
따라서 초반 layer에서는 Maxpooling을 사용하지 않고, (3,3)filter사용과 함께 채널을 16,32,64로 늘림

CNN layer가 끝난 후에는 Flatten을 통해 1D 벡터로 만든 후 Dense층을 통과.

마지막 Dense는 10개의 Label에 맞추어 10개로 할당, activation은 softmax

```python
import tensorflow as tf
import keras 
from keras import layers
from keras import models
from keras.layers import *
model1 = models.Sequential()
model1.add(layers.Conv2D(16,(3,3),activation='relu',input_shape = (28,28,1),name='conv2d_layer1'))
model1.add(layers.Conv2D(32,(3,3),activation='relu'))
model1.add(layers.Conv2D(64,(3,3),activation='relu'))
model1.add(layers.MaxPooling2D((2,2)))
model1.add(layers.BatchNormalization(
    axis=-1, momentum=0.99, epsilon=0.001))
model1.add(layers.Conv2D(128,(3,3),activation='relu',name='conv2d_layer2'))
model1.add(layers.MaxPooling2D((2,2)))
model1.add(layers.BatchNormalization(
    axis=-1, momentum=0.99, epsilon=0.001))
model1.add(layers.Flatten())
model1.add(Dense(64,activation = 'relu'))
model1.add(Dense(32,activation = 'relu'))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

 learning_rate를 0.1로 했을 때는 잘 수렴하지 않음(성능저하)

```python
Adam=tf.keras.optimizers.Adam(
    learning_rate=0.01,epsilon=1e-1, amsgrad=False,name='Adam')
model1.compile(
    optimizer=Adam,
    loss='categorical_crossentropy',
    metrics=['accuracy'])

y_pred_tr=model1.fit(X_train_ovr,y_tr_lb,epochs=1,batch_size=32,validation_split=0.2)
```

```python
try:
    plt.plot(y_pred_tr.history['accuracy'])
    plt.plot(y_pred_tr.history['val_accuracy'])
except KeyError:
 plt.plot(y_pred_tr.history['accuracy'])
 plt.plot(y_pred_tr.history['val_accuracy'])
 plt.title('Accuracy vs. epochs')
 plt.ylabel('Loss')
 plt.xlabel('Epoch')
 plt.legend(['Training'], loc='lower right')
 plt.show() 
```

```python
plt.plot(y_pred_tr.history['loss'])
plt.plot(y_pred_tr.history['val_loss'])
plt.title('Loss vs. epochs')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show() 
```
