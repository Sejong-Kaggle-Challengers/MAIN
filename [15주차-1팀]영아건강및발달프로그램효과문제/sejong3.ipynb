```python
submit=pd.read_csv('/kaggle/input/sejong-ai-challenge-p3/sample_submission.csv')
train=pd.read_csv('/kaggle/input/sejong-ai-challenge-p3/train.csv')
test=pd.read_csv('/kaggle/input/sejong-ai-challenge-p3/test.csv')
```

```python
!pip install pycaret
```

```python
from pycaret.regression import *

reg=setup(session_id=71,data=train,target='y',train_size=0.8)
```

```python
models()
```

```python
best3=compare_models(exclude=['xgboost','catboost','lar'],n_select=3)
```

```python
# blend_best3=blend_models(best3)

omp=create_model('omp')
stack_best3=stack_models(best3,meta_model=omp)
```

```python
final_stack=finalize_model(stack_best3)
pred=predict_model(final_stack,data=test)
```

```python
submit['y']=pred['Label']
submit.head()
```
