```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")

train
```

```python
train['mom.edu'].unique()
```

```python
train.isna().sum()
```

```python
train=train.drop(labels='dadage',axis=1)
test=test.drop(labels='dadage',axis=1)
```

```python
train['mom.edu']=train['mom.edu'].fillna(train['mom.edu'].mean())
```

```python
train['site']=train['site'].fillna(train['site'].mean())
```

```python
test['mom.edu']=test['mom.edu'].fillna(test['mom.edu'].mean())
test['site']=test['site'].fillna(test['site'].mean())
```

```python
train=train.drop(labels='id',axis=1)
test=test.drop(labels='id',axis=1)
```

```python
!pip install autogluon
```

```python
from autogluon.tabular import TabularPredictor
```

```python
label = 'y'
eval_metric = 'mean_squared_error'  
```

```python
predictor = TabularPredictor(label=label, eval_metric=eval_metric, verbosity=3).fit(
    train, presets='best_quality', time_limit=3600
)
```

```python
results = predictor.fit_summary()
```

```python
prediction=predictor.predict(test)
```

```python
submission=pd.read_csv("sample_submission.csv")
```

```python
submission
```

```python
for i in range(len(submission)):
    submission['y']=prediction
submission
```

```python
submission.to_csv("result.csv",index=False)

```
